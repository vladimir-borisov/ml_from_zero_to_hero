```python
from mlzero.supervised_learning.linear_regression import LinearRegression, LinearRegressionSGD
from sklearn.linear_model import LinearRegression as LinearRegressionSklearn
from sklearn.linear_model import SGDRegressor
from sklearn.datasets import make_regression
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings("ignore")
```

```python
X, y = make_regression(n_samples=500, n_features=1, noise=3)
```

# Linear Regression (mlzero)

```python
model = LinearRegression().fit(X, y)
print(model.weights)
sns.lineplot(X[:, 0], y)
sns.lineplot(X[:, 0], model.predict(X.copy()))
```

# Linear Regression (sklearn)

```python
model_sklearn = LinearRegressionSklearn().fit(X, y)
print(model_sklearn.coef_, model_sklearn.intercept_)
sns.lineplot(X[:, 0], y)
sns.lineplot(X[:, 0], model_sklearn.predict(X))
```

# Linear Regression + SGD (mlzero)

```python
model_sgd = LinearRegressionSGD(max_iter=100, learning_rate = 0.001).fit(X, y)

print(model_sgd.weights)
sns.lineplot(X[:, 0], y)
sns.lineplot(X[:, 0], model_sgd.predict(X))
```

# Linear Regression + SGB (sklearn)

```python
model_sgd_sklearn = SGDRegressor(max_iter=1001).fit(X, y)

print(model_sgd.weights)
sns.lineplot(X[:, 0], y)
sns.lineplot(X[:, 0], model_sgd.predict(X))
```
