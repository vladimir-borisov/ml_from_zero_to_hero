```python
import numpy as np
from sklearn.datasets import make_regression
from mlzero.supervised_learning.linear_regression import LinearRegression
import seaborn as sns
import warnings
warnings.filterwarnings("ignore")
```

```python
X, y = make_regression(n_samples=500, n_features=1, noise=3)
```

```python
model = LinearRegression().fit(X, y)
```

```python
sns.lineplot(X[:, 0], y)
sns.lineplot(X[:, 0], model.predict(X))
```
