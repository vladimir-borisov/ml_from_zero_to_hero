```python
from mlzero.supervised_learning.linear_regression import LinearRegression, LinearRegressionSGD, LassoRegression
from sklearn.linear_model import LinearRegression as LinearRegressionSklearn
from sklearn.linear_model import SGDRegressor, Lasso
from sklearn.datasets import make_regression
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
import warnings
warnings.filterwarnings("ignore")
```

```python
X, y = make_regression(n_samples=500, n_features=1, noise=30)
```

# Linear Regression (Ordinary Least Squares)

```python
model_mlzero = LinearRegression().fit(X, y)
model_sklearn = LinearRegressionSklearn().fit(X, y)


plt.figure(figsize=(12, 5), dpi = 100)


plt.subplot(1, 2, 1)
plt.title(f'Linear Regression (mlzero)\n'
          f'coefficient={round(model_mlzero.weights[1], 2)}, bias={round(model_mlzero.weights[0], 2)}')
sns.lineplot(X[:, 0], y)
sns.lineplot(X[:, 0], model_mlzero.predict(X.copy()))


plt.subplot(1, 2, 2)
plt.title(f'Linear Regression (sklearn)\n'
          f'coefficient={round(model_sklearn.coef_[0], 2)}, bias={round(model_sklearn.intercept_, 2)}')
sns.lineplot(X[:, 0], y)
sns.lineplot(X[:, 0], model_sklearn.predict(X))
```

# Linear Regression (Stochastic Gradient Descent)

```python
model_sgd_mlzero = LinearRegressionSGD(max_iter=100, learning_rate = 0.001).fit(X, y)
model_sgd_sklearn = SGDRegressor(max_iter=100, learning_rate = 'constant', eta0 = 0.001).fit(X, y)



plt.figure(figsize=(12, 5), dpi = 100)


plt.subplot(1, 2, 1)
plt.title(f'Linear Regression (SGD) (mlzero)\n'
          f'coefficient={round(model_sgd_mlzero.weights[1], 2)}, bias={round(model_sgd_mlzero.weights[0], 2)}')
sns.lineplot(X[:, 0], y)
sns.lineplot(X[:, 0], model_sgd_mlzero.predict(X))


plt.subplot(1, 2, 2)
plt.title(f'Linear Regression (SGD) (sklearn)\n'
          f'coefficient={np.round(model_sgd_sklearn.coef_[0], 2)}, bias={np.round(model_sgd_sklearn.intercept_[0], 2)}')
sns.lineplot(X[:, 0], y)
sns.lineplot(X[:, 0], model_sgd_sklearn.predict(X))
```

# Lasso Regression (L1 Norm)

```python
model_lasso_mlzero = LassoRegression(max_iter = 100, learning_rate = 0.001).fit(X, y)
model_lasso_sklearn = Lasso(max_iter = 100).fit(X, y)

plt.figure(figsize=(12, 5), dpi = 100)


plt.subplot(1, 2, 1)
plt.title(f'Lasso Regression (mlzero)\n'
          f'coefficient={round(model_lasso_mlzero.weights[1], 2)}, bias={round(model_lasso_mlzero.weights[0], 2)}')
sns.lineplot(X[:, 0], y)
sns.lineplot(X[:, 0], model_lasso_mlzero.predict(X))


plt.subplot(1, 2, 2)
plt.title(f'Lasso Regression (sklearn)\n'
          f'coefficient={np.round(model_lasso_sklearn.coef_[0], 2)}, bias={np.round(model_lasso_sklearn.intercept_, 2)}')
sns.lineplot(X[:, 0], y)
sns.lineplot(X[:, 0], model_lasso_sklearn.predict(X))
```
